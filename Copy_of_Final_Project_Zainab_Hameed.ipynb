# **Task C: Data Preprocessing for Modeling**


C1. Feature Engineering: Create new features if relevant

```python
# Feature 1: Extract year, month, week
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Week'] = df['Date'].dt.isocalendar().week

# Feature 2: Sales difference from previous week
df['Sales_Diff'] = df['Weekly_Sales'].diff().fillna(0)


# Feature 4: Fuel price change from previous week
df['Fuel_Change'] = df['Fuel_Price'].diff().fillna(0)

# Feature 5: Temperature bin (cold, mild, warm, hot)
df['Temp_Bin'] = pd.cut(df['Temperature'], bins=[-10, 40, 60, 80, 100],
                        labels=['Cold', 'Mild', 'Warm', 'Hot'])
print(df.head())
```

**Explanation**
This program adds new features to a dataset to help analyze trends. It extracts the year, month, and week from a date, calculates weekly changes in sales and fuel prices, and categorizes temperatures into bins like "Cold" or "Hot." These features make the data easier to understand and more useful for analysis or modeling.

C2. Scaling/Normalization: Scale or normalize numerical data if necessary.

```python
from scipy.stats import zscore
# Step 1: Calculate Z-scores
df['Z_score'] = zscore(df['Weekly_Sales'])

# Step 2: Identify outliers (Z > 2 or Z < -2)
outliers = df[(df['Z_score'] > 2) | (df['Z_score'] < -2)]
mean_weekly_sales = df['Weekly_Sales'].mean()
print("\n Mean :",mean_weekly_sales)
print("\n Detected Outliers:")
print(outliers)
```

**Explanation**

Each sales value is converted into a Z-score to show how unusual it is compared to the average.A Z-score of 0 means it’s exactly average; ±2 or more means it's far from average.

**D4. Dash Application for Web-based Interaction**


Build a Dash dashboard to allow users to interact with visual data.


Include dynamic elements like dropdown filters, sliders, and interactive graphs.


Enable viewing of:


Distributions of variables


Relationships between key features


Interactive correlation analysis


Category breakdowns and trends


```python
!pip install dash
import dash
from dash import dcc, html, Input, Output
import plotly.express as px
import pandas as pd
import numpy as np
df=pd.read_csv('/content/Walmart_Store_sales - Walmart_Store_sales.csv')
df= pd.DataFrame(df)
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# Start Dash app
app = dash.Dash(__name__)

numerical_columns = ['Weekly_Sales', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']

app.layout = html.Div([
    html.H1(" Interactive Sales Dashboard", style={'textAlign': 'center'}),

    html.Div([
        html.Label("Select Variable for Distribution"),
        dcc.Dropdown(
            id='dist-var',
            options=[{'label': col, 'value': col} for col in numerical_columns],
            value='Weekly_Sales'
        ),
        dcc.Graph(id='histogram')
    ], style={'width': '48%', 'display': 'inline-block'}),

    html.Div([
        html.Label("Select X and Y for Relationship Plot"),
        dcc.Dropdown(
            id='x-var',
            options=[{'label': col, 'value': col} for col in numerical_columns],
            value='Fuel_Price'
        ),
        dcc.Dropdown(
            id='y-var',
            options=[{'label': col, 'value': col} for col in numerical_columns],
            value='Weekly_Sales'
        ),
        dcc.Graph(id='scatter-plot')
    ], style={'width': '48%', 'display': 'inline-block', 'float': 'right'}),

    html.H2("📈 Correlation Heatmap"),
    dcc.Graph(
        id='correlation-heatmap',
        figure=px.imshow(df[numerical_columns].corr(),
                         text_auto=True,
                         aspect="auto",
                         title="Correlation Matrix")
    )
])

# Callbacks
@app.callback(
    Output('histogram', 'figure'),
    Input('dist-var', 'value')
)
def update_histogram(selected_var):
    fig = px.histogram(df, x=selected_var, nbins=10,
                       title=f'Distribution of {selected_var}')
    return fig

@app.callback(
    Output('scatter-plot', 'figure'),
    [Input('x-var', 'value'),
     Input('y-var', 'value')]
)
def update_scatter(x, y):
    fig = px.scatter(df, x=x, y=y, color='Holiday_Flag',
                     hover_name='Date',
                     title=f'{y} vs {x}')
    return fig

# Run the app
if __name__ == '__main__':
    app.run(debug=True)
```
